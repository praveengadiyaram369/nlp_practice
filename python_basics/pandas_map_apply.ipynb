# Loading Auto MPG data

```
import os
import pandas as pd
import numpy as np

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", 
    na_values=['NA', '?'])

display(df)
```

# Map method in pandas

1.  map method is used to transform the values of a column and works only on single column.
2.  this method expects the mapping is provided in the form of a dictionary.
3.  keys of the dictionary is mapped to values of the target column
4.  finally the output of the method map results in another list of values representing a new column

```
df['origin_continent'] = df['origin'].map({1: 'America', 2: 'Europe', 3: 'Asia'})
display(df)
```

# Apply method in pandas

1.  apply method can run over all the data of a dataframe and the data can be provided to either a named function or to a lambda function.
2.  the output of apply function is same as map, list of values representing a new column(same no. of row length).
3.  map is more similar to replce, but apply can achieve more complex transformed values.

```
df['efficiency'] = df.apply(lambda x: x['displacement']/x['horsepower'], axis=1)
display(df)
```

```
import pandas as pd

df=pd.read_csv('https://www.irs.gov/pub/irs-soi/16zpallagi.csv')
```

```
df = df[['STATE','zipcode','agi_stub','N1']]
display(df)
```

```
df = df.loc[(df['zipcode'] != 0) & (df['zipcode'] != 99999)]
display(df)
```

```
medians = {1:12500,2:37500,3:62500,4:87500,5:112500,6:212500}
df['agi_stub'] = df['agi_stub'].map(medians)
display(df)
```

```
groups = df.groupby(by='zipcode')
display(groups)
```

```
df = pd.DataFrame(groups.apply(lambda x: (x['agi_stub']+x['N1'])))
display(df)
```
