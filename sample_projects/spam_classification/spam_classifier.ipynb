```python
import os
import nltk
import pandas as pd

import re 
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
input_file_path = os.getcwd() + '/smsspamcollection/SMSSpamCollection'
df = pd.read_csv(input_file_path, sep='\t', names=['label', 'text'])
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    
    data = re.sub('[^a-zA-Z]', ' ', text)
    data = data.lower()
    data = data.split()
    data = [lemmatizer.lemmatize(word) for word in data if word not in stop_words]
    
    return ' '.join(data)
```

```python
def get_accuracy(X, y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    model = RandomForestClassifier(n_estimators=100).fit(X_train, y_train)
    preds = model.predict(X_test)

    accuracy = accuracy_score(y_test, preds)
    print(f'Accuracy {accuracy*100}%')
```

```python
df['text_cleaned'] = df['text'].apply(lambda x: clean_text(x))
df.head()
```

```python
corpus = df.text_cleaned.values
corpus
```

```python
y = list(df['label'].apply(lambda x: 1 if x == 'ham' else 0))
y
```

```python
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
get_accuracy(X, y)
```

```python
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(corpus).toarray()
get_accuracy(X, y)
```
